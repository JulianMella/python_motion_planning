```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data_frame = pd.read_csv('file1.csv')

# Clean up dataset
for x in data_frame.index:
    if data_frame.loc[x, "cost"] == '[]':
        data_frame.drop(x, inplace=True)


data_frame["cost"] = data_frame["cost"].astype(float)

short_radius_data = data_frame[data_frame["radius"] == "short"]
sns.boxplot(
    data=short_radius_data,
    x="algorithm",
    y="cost",
    hue="algorithm",
    medianprops={"color": "r", "linewidth": 2},
    fill=False
)

plt.title("Short radius (15)")
plt.xlabel("Algorithm")
plt.ylabel("Cost")

plt.show()
```

```python

long_radius_data = data_frame[data_frame["radius"] == "long"]

short_radius_data = data_frame[data_frame["radius"] == "short"]
sns.boxplot(
    data=long_radius_data,
    x="algorithm",
    y="cost",
    hue="algorithm",
    medianprops={"color": "r", "linewidth": 2},
    fill=False
)

plt.title("Short radius(15)")
plt.xlabel("Algorithm")
plt.ylabel("Cost")

plt.show()
```

```python
#Create graph that shows how computational cost is affected by the cost to target 
sns.scatterplot(
    data=data_frame,
    x="cost",
    y="search_area",
    hue="algorithm"
)
plt.title("Computational Cost vs Cost to Target")
plt.xlabel("Cost to Target")
plt.ylabel("Computational Cost (search_area)")
plt.show()
```
